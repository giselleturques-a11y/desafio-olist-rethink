```python
from pyspark.sql.functions import current_timestamp

# Usando exatamente o formato que o display(dbutils.fs.ls) te mostrou
path_input = "dbfs:/Volumes/workspace/default/data/raw/"
path_output = "dbfs:/Volumes/workspace/default/data/delta/bronze/"

# IMPORTANTE: Verifique se os nomes na lista abaixo batem 100% com os arquivos no Volume
# (Ex: se no volume estiver 'Orders.csv', mude aqui para 'Orders')
tabelas = ["olist_customers_dataset", "olist_geolocation_dataset", "olist_order_items_dataset", "olist_order_payments_dataset", "olist_order_reviews_dataset", "olist_orders_dataset", "olist_products_dataset", "olist_sellers_dataset", "product_category_name_translation"]

for nome in tabelas:
    try:
        full_path = f"{path_input}{nome}.csv"
        print(f"Lendo: {full_path}")
        
        # Lendo o CSV
        df = spark.read.format("csv") \
            .option("header", "true") \
            .option("inferSchema", "true") \
            .load(full_path)
        
        # 2.3.1 - Adicionando timestamp
        df_bronze = df.withColumn("ingestion_timestamp", current_timestamp())
        
        # 2.3.2 - Salvando em Delta
        # Criamos a pasta bronze automaticamente
        df_bronze.write.format("delta") \
            .mode("overwrite") \
            .save(f"{path_output}{nome}")
            
        print(f"✅ Tabela {nome} carregada com sucesso!")
        
    except Exception as e:
        print(f"❌ Erro na tabela {nome}: Verifique se o nome do arquivo está correto (maiúsculas/minúsculas).")
        # print(f"Detalhe do erro: {e}") # Descomente se precisar ver o erro técnico
```
