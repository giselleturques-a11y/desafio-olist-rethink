```python
# 1. Configuração de caminhos
path_gold = "dbfs:/Volumes/workspace/default/data/delta/gold/"
path_output = "dbfs:/Volumes/workspace/default/data/delta/output/"

# Lista das tabelas Gold para exportação
gold_tables = ["customer_summary", "product_summary", "seller_summary"]

print(f"🚀 Iniciando simulação de partilha de dados (Delta Sharing)...")

for table in gold_tables:
    try:
        print(f"📥 A ler tabela Gold: {table}")
        
        # Ler os dados da camada Gold
        df = spark.read.format("delta").load(f"{path_gold}{table}")
        
        # Definir o caminho de exportação CSV
        export_path = f"{path_output}gold_{table}_export.csv"
        
        # Exportar para CSV
        # .coalesce(1) garante que o resultado seja um único arquivo CSV
        # .header(True) inclui os nomes das colunas no ficheiro
        df.coalesce(1).write.format("csv") \
            .mode("overwrite") \
            .option("header", "true") \
            .option("delimiter", ",") \
            .save(export_path)
            
        print(f"✅ Exportação concluída: {export_path}")
        
    except Exception as e:
        print(f"❌ Erro ao exportar a tabela {table}: {str(e)}")

print(f"🏁 Simulação finalizada. Os dados estão prontos para consumo externo no diretório /output/.")
```
